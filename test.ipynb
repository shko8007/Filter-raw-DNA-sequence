```python
read_file_name='C:/Users/kos3/Downloads/final project_2019_python/input_data_516_visit1.fasta'
```

```python
raw_data=open(read_file_name,"r")
```

```python
a=raw_data.read()
```

```python
print(a)
```

```python
b=a.replace('\n','') # merge all separated lines
c=b.split('>') # split into individual sequence based on sequence ID
d='\n'.join(c) # convet list to string
e=d.replace('\n','',1).replace('m','>m') # remove space and add additional characters
f=e.split()
```

```python
print(f)
```

```python
x_mod='' # initialize x_mod
for x in f:
    value_A=x.count("A")
    value_G=x.count("G")
    value_C=x.count("C")
    value_T=x.count("T")
    total_value=value_A+value_G+value_C+value_T # Sum of bases
    
    if total_value > 2500 and total_value < 3000 : # Size criteria to filter off-target
        
        x_mod = x_mod + x
    else:
        continue
```

```python
print(x_mod)
```

```python
h=x_mod.split('>')
i='\n'.join(h)
j=i.replace('\n','',1).replace('m','>m').replace('s','s\n')
```

```python
print(j)
```

```python
writing_file_name='output.txt'
mod_data=open(writing_file_name, 'w')
mod_data.write(j)
mod_data.close()
```
